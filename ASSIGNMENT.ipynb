```python
np.random.seed(101)



num_ports = 5000
all_weights = np.zeros((num_ports, len(returns.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for ind in range(num_ports):
    
    weights = np.array(np.random.random(20))
    weights = weights/np.sum(weights)
    
    #save weights
    all_weights[ind,:] =weights 
    
    #Expected Return
    ret_arr[ind]= np.sum((log_ret.mean()*weights) * 252)
    
    #Expected Volatility
    
    vol_arr[ind] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov()*252,weights)))
    
    #Sharpe Ratio
    sharpe_arr[ind] = ret_arr[ind]/vol_arr[ind]
    
    
    
```

The value of maximum sharpe Ratio

```python
sharpe_arr.max()
```

The index position of maximum sharpe ratio

```python
sharpe_arr.argmax()
```

The volatility for maximum sharpe ratio

```python
max_sr_vol = vol_arr[1607]
max_sr_vol
```

The returns for maximum sharpe ratio

```python
max_sr_ret = ret_arr[1607]
max_sr_ret
```

The weights of individual securities as suggested by Monte Carlo Simulation

```python
d = {}
for i in range(20):
    d[list(returns.columns)[i]] = list(all_weights[1607, :])[i]    
```

```python
s = pd.Series(d)
s.reset_index()
df = pd.DataFrame(s)
```
